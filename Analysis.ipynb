```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('flights.csv')
dataByAirline = data.groupby("AIRLINE")
print(dataByAirline['ARRIVAL_DELAY'].apply(lambda delays: delays[delays>0].mean()))
print(dataByAirline['ARRIVAL_DELAY'].apply(lambda delays: delays[delays<0].mean()))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('flights.csv')
dataByAirline = data.groupby("AIRLINE")
print(dataByAirline['ARRIVAL_DELAY'].apply(lambda delays: delays[delays>0].sum()))
print(dataByAirline['ARRIVAL_DELAY'].count())
print(dataByAirline['ARRIVAL_DELAY'].apply(lambda delays: delays[delays<0].sum()))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', None)
data = pd.read_csv('flights.csv')

dataByAirline = data.groupby("ORIGIN_AIRPORT")
print(dataByAirline['ARRIVAL_DELAY'].apply(lambda delays: delays[delays>0].mean()).sort_values(ascending=False))
dataByAirline['ARRIVAL_DELAY'].apply(lambda delays: delays[delays<0].mean()).sort_values()
```
