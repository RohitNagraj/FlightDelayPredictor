```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('flights.csv')
```

# AIRLINE MEAN

```python
dataByAirline = data.groupby("AIRLINE")
print(dataByAirline['ARRIVAL_DELAY'].apply(lambda delays: delays[delays>0].mean()).sort_values(ascending=False))
print(dataByAirline['ARRIVAL_DELAY'].apply(lambda delays: delays[delays<0].mean()).sort_values())
```

# AIRLINE SUM AND FREQUENCY

```python
print(dataByAirline['ARRIVAL_DELAY'].apply(lambda delays: pd.Series(data=[  delays[delays>0].sum()  ,  delays[delays>0].count()  ],index=["Total Delay","Frequency"])))
print(dataByAirline['ARRIVAL_DELAY'].apply(lambda delays: pd.Series(data=[  delays[delays<0].sum()  ,  delays[delays<0].count()  ],index=["Total Advance","Frequency"])))
```

```python
dataByDay = data.groupby("DAY_OF_WEEK")
print(dataByDay['ARRIVAL_DELAY'].apply(lambda delays: delays[delays>0].mean()).sort_values(ascending=False))
print(dataByDay['ARRIVAL_DELAY'].apply(lambda delays: delays[delays<0].mean()).sort_values())
```
